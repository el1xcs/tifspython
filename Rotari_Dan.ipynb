```python
#62
"""
    Printează o piramidă cu înălțimea n folosind caracterul "*".

    Parametri:
    - n (int): Numărul întreg reprezentând înălțimea piramidei.

    Exemplu de utilizare:
    >>> print_pyramid(4)
       *
      ***
     *****
    *******
"""

def print_pyramid(n):
    for pop in range(1, n + 1):
        space = " " * (n - pop)
        star = "*" * (2 * pop - 1)
        print(space + star)

print_pyramid(4)

#63
"""
    Verifică dacă un număr este prim.

    Parametri:
    - num (int): Numărul întreg de verificat.

    Returnează:
    - bool: True dacă numărul este prim, False în caz contrar.
"""
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(19))  
print(is_prime(30))  

#64
"""
    Uneste două liste sortate și returnează o nouă listă sortată.

    Parametri:
    - list1 (list): Prima listă sortată.
    - list2 (list): A doua listă sortată.

    Returnează:
    - list: O nouă listă sortată care conține toate elementele din list1 și list2.


"""
def merge_lists(list1, list2):

    result = []
    ilo = joi = 0

    while ilo < len(list1) and joi < len(list2):
        if list1[ilo] < list2[joi]:
            result.append(list1[ilo])
            ilo += 1
        else:
            result.append(list2[joi])
            joi += 1

    result.extend(list1[ilo:])
    result.extend(list2[joi:])

    return result

print(merge_lists([1, 3, 5,8], [2, 4, 6,7]))

#65
from statistics import median, mode

def calculate_stats(numere):
    """
    Calculează media, mediana și modul unei liste de numere.

    Parametri:
    - numere (list): Lista de numere pentru care se calculează statisticile.

    Returnează:
    - dict: Un dicționar conținând media (cheia 'media'), mediana (cheia 'mediana') și modulul (cheia 'modul').
            Dacă există mai multe moduri, 'modul' va fi o listă cu toate valorile modale.
"""

    if not numere:
        raise ValueError("Lista trebuie să conțină cel puțin un număr.")

    rezultate = {
        'media': sum(numere) / len(numere),
        'mediana': median(numere),
        'modul': mode(numere)
    }

    return rezultate


print(calculate_stats([1, 2, 2,2,7,9,20,59]))

```

```python
#66
def fibonacci(n):
    """
    Returnează al n-lea element din șirul lui Fibonacci.

    Parametri:
    - n (int): Poziția din șirul lui Fibonacci pentru care se dorește calculul elementului.

    Returnează:
    - int: Valoarea al n-lea element din șirul lui Fibonacci.

    """

    fib_list = [0, 1]

    while len(fib_list) <= n:
        fib_list.append(fib_list[-1] + fib_list[-2])

    return fib_list[n]

print(fibonacci(int(input('Nr'))))
```

```python
#67
def invert_dict(dictionar):
    """
    Inversează cheile cu valorile într-un dicționar.

    Parametri:
    - dictionar (dict): Dicționarul inițial care va fi inversat.

    Returnează:
    - dict: Un nou dicționar cu cheile și valorile inversate.
            Valorile sunt stocate sub formă de listă în cazul în care există mai multe chei cu aceeași valoare.

    """

    inverted_dict = {}

    for chei, valoare in dictionar.items():
        if valoare not in inverted_dict:
            inverted_dict[valoare] = [chei]
        else:
            inverted_dict[valoare].append(chei)

    return inverted_dict

print(invert_dict({'a': 1, 'b': 2, 'c': 1}))

#68

def char_histogram(sir):
    """
    Calculează frecvența fiecărui caracter dintr-un șir și returnează un dicționar.

    Parametri:
    - sir (str): Șirul de caractere pentru care se dorește calculul frecvenței.

    Returnează:
    - dict: Un dicționar cu caracterele din șir și frecvența fiecărui caracter.

    """

    histograma = {}

    for caract in sir:
        if caract in histograma:
            histograma[caract] += 1
        else:
            histograma[caract] = 1

    return histograma

print(char_histogram("abracadabra"))
    
#69
def unique_elements(lista):
    """
    Extrage elementele unice dintr-o listă și returnează o nouă listă.

    Parametri:
    - lista (list): Lista din care se extrag elementele unice.

    Returnează:
    - list: O nouă listă conținând doar elementele unice din lista dată.

    """

    lista_unica = []

    for element in lista:
        if element not in lista_unica:
            lista_unica.append(element)

    return lista_unica


print(unique_elements([1, 2, 2, 3, 4, 4, 5]))

#70
def common_elements(lista1, lista2):
    """
    Returnează o nouă listă cu elementele comune ale celor două liste.

    Parametri:
    - lista1 (list): Prima listă de comparat.
    - lista2 (list): A doua listă de comparat.

    Returnează:
    - list: O nouă listă conținând elementele comune ale celor două liste.

    """

    lista_comuna = [element for element in lista1 if element in lista2]

    return lista_comuna

print(common_elements([1, 2, 3, 4], [3, 4, 5,]))
```
